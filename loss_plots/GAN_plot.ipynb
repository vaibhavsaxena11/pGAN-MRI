```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
def read_file(filename):
    file1 = open(filename, 'r')
    lines = file1.readlines()
    dloss = []
    gloss = []
    gper = []
    gmse = []
    gfreq = []
    time = []
    for i in range(len(lines)):
        # print(i)
        l = (lines[i][:-3]).split(' ')
        # print(l)
        dloss.append(float(l[3]))
        gloss.append(float(l[5]))
        gper.append(float(l[7]))
        gmse.append(float(l[9]))
        gfreq.append(float(l[11]))
        time.append(float(l[13]))
    file1.close()
    return np.array(dloss), np.array(gloss), np.array(gper), np.array(gmse), np.array(gfreq), np.array(time)
    
def plot_loss(loss, title):
    sns.lineplot(range(1, len(loss)+1), loss)
    plt.ylabel(title)
    plt.xlabel("Step")
    plt.show()
```

```python
file_list = ['epoch_1.txt', 'epoch_5.txt', 'epoch_10.txt']
for f_name in file_list:
    print("Plots for ", f_name[:-4])
    dloss, gloss, gper, gmse, gfreq, time = read_file(f_name)
    print()
    print("Discriminator Loss")
    plot_loss(dloss, "Discriminator Loss")
    print("Combined Generator Loss")
    plot_loss(gloss, "Combined Generator Loss")
    print("Perceptual Generator Loss")
    plot_loss(gper, "Perceptual Generator Loss")
    print("MSE Generator Loss")
    plot_loss(gmse, "MSE Generator Loss")
    print("Frequency Generator Loss")
    plot_loss(gfreq, "Frequency Generator Loss")
```
